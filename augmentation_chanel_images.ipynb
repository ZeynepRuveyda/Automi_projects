```python
image = cv2.imread('/content/drive/My Drive/Exxelia/c.png')
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
lower_white = np.array([0, 0, 0], dtype=np.uint8)
upper_white = np.array([255, 255, 255], dtype=np.uint8)
mask = cv2.inRange(resized, lower_white, upper_white) # could also use threshold
res = cv2.bitwise_not(resized, resized, mask)

cv2_imshow(res) # gives black background

cv2.imwrite('/content/drive/My Drive/Exxelia/gray.png',res)
```

Augmentation

```python
#rotate images

for i in range(1,13):

  
  img = cv2.imread('/content/drive/My Drive/chanel_photos/good/g_'+str(i)+'.png')
  print(type(img))
  # <class 'numpy.ndarray'>

  print(img.shape)
  #rotate

  img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
  img_rotate_90_counterclockwise = cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
  img_rotate_180 = cv2.rotate(img,cv2.ROTATE_180)

  #flip
  
  img_flip_ud = cv2.flip(img, 0)

  img_flip_lr = cv2.flip(img, 1)

  img_flip_ud_lr = cv2.flip(img, -1)

  #save
  
  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_rotate_.png', img_rotate_90_clockwise)

  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_counterclockwise_.png', img_rotate_90_counterclockwise)

  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_rotate_180_.png', img_rotate_180)
  
  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_flip_ud_.png', img_flip_ud)

  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_flip_lr_.png', img_flip_lr)

  cv2.imwrite('/content/drive/My Drive/chanel_photos/good/'+str(i)+'_flip_ud_lr_mask.png', img_flip_ud_lr)
```
